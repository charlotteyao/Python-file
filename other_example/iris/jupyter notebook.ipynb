# Matplotlib集成
如果你用python绘制过图形，那你肯定知道matplotlib。Matplotlib是一个用于创建漂亮图形的Python库，结合Jupyter notebook使用时体验更佳。

要想在Jupyter notebook中使用matplotlib，需要告诉Jupyter获取生成的所有图形，并将其嵌入notebook中。为此，需要计算：

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(20)
y = x**2

plt.plot(x,y)
```
